__5*.__ Написать на Python функцию для перемножения двух произвольных матриц, не используя NumPy.

```python
def matrix_multiplication(A, B):
    result = np.zeros((len(A), len(B[0])))
    try:
        if not len(A) == len(B[0]) and not len(A) == len(A[0]):
            raise IndexError
        for line in range(len(A)):
            for value in range(len(B[0])):
                result[line][value] += sum(
                    [A[line][pos] * B[pos][value] for pos in range(len(A[0]))])
        return print(*result, sep = '\n')
    except IndexError:
        print('Произведение матриц не определено')
```
