__5*.__ Написать на Python программу с реализацией одного из изученных алгоритмов решения СЛАУ.

```python
import numpy as np

def minor(M, c):
    B = [[1] * len(M) for i in range(len(M))]
    for l in range(len(M)):
        for k in range(len(M)):
            B[l][k] = M[l][k]
    
    B.pop(0)
    
    for i in range(len(B)):
        B[i].pop(c)
    return B


def det(M):
    X = 0
    if len(M) != len(M[0]):
        print('Matrix is not square')
    else:
        if len(M)<=2:
            return M[0][0] * M[1][1] - M[0][1] * M[1][0]
        else:
            for i in range(len(M)):
                X = X + ((-1)**i) * M[0][i] * det(minor(M, i))
        return X

    
def cramer(A,B):
    det_A = det(A)
    X = []
    if det_A == 0:
        print('The determinant of a matrix must not be zero')
        return X
    n = len(A)    
    for i in range(n):
        matrix = A.copy()
        matrix[:, i] = B
        print('\n', matrix)
        X.append(det(matrix) / det_A)
    return X
```

```python
a = np.array([[1,-2], [3,-4]])
b = np.array([1, 7])
print(f'Матрица коэффициентов:\n{a}')
print('\n',b)
print('\nОпределитель:', det(a))
print('\nОтвет:', cramer(a, b))
```
